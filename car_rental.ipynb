```python
import datetime
import time
```

```python
class renting_Cars():
    
    def __init__(self,cars_available):
        self.cars_available = cars_available;
        
    def available_Cars(self,cars_needed):
    
        # print(self.cars_available)
        if self.cars_available < cars_needed:
            print(f' Only {self.cars_available} are available')
            return 1;
        else :
            print(f" Congrats {cars_needed} cars are available");
            self.cars_available -=  cars_needed;
            # print(self.cars_available)
            return 2;
            
    def weekly(self):
         duration = int(input('Enter Week duration from now:'))
         start_time = datetime.datetime.now()
         str_time = start_time.strftime("%b %d %Y %I:%M%p")
         print(f"Kindly note your start time : \n {str_time} \n HAVE A SAFE JOURNEY..!")
         return start_time;
         
    def hourly(self):
        duration = int(input('Enter Hour duration from now:'))
        start_time = datetime.datetime.now()
        str_time = start_time.strftime("%b %d %Y %I:%M%p")
        print(f"Kindly note your start time : \n {str_time} \n HAVE A SAFE JOURNEY..!")
        return start_time;
        
        
    def daily(self):
         duration = int(input('Enter Day duration from now:'))
         start_time = datetime.datetime.now()
         str_time = start_time.strftime("%b %d %Y %I:%M%p")
         print(f"Kindly note your start time : \n {str_time} \n HAVE A SAFE JOURNEY..!")
         return start_time;

    def car_Return(self,rental_type):
            
            start_date=input("Enter the start Date and Time of your rental \n (Ex-Jan 1 2000 12:00AM)")
            format = '%b %d %Y %I:%M%p'
            rental_start_time = datetime.datetime.strptime(start_date, format)
             
            if rental_start_time > datetime.datetime.now():
                print("Enter Valid Date")
            else:
                current_time = datetime.datetime.now()
                Total_rental_time = current_time - rental_start_time
                # inbulit function used to convert day and hrs into hrs format
                deltatime=datetime.timedelta(days=Total_rental_time.days,seconds=Total_rental_time.seconds)
                hours=float(deltatime.days * 24 + deltatime.seconds // 3600)
              
                rented_Cars = int(input("No of Cars Rented: "))
                self.cars_available += rented_Cars
                print(self.cars_available)
            if rental_type == 1:
                # HOURLY RENTAL CHARGES
                 
                 Rent = 100 * hours *  rented_Cars
                 print(f" Your Total Rental Hours: {hours} \n Bill Amount : {Rent}")
            elif rental_type == 2:
                # DAILY RENTAL CHARGES
                 Rent = 90 * hours *  rented_Cars
                 print(f" Your Total Rental Hours: {hours} \n Bill Amount : {Rent}") 
            elif rental_type == 3:
                # WEEKLY RENTAL CHARGES
                 Rent = 85 * hours *  rented_Cars
                 print(f" Your Total Rental Hours: {hours} \n Bill Amount : {Rent}")
            return Rent;


```

```python

class customer(renting_Cars):
    def __init__(self,cars_available):
        super().__init__(cars_available)
        
    def request_Cars(self,cars):
       # print("in child fun")
       car_avai=super().available_Cars(cars)
       if car_avai == 2:
           print("Select Your Rental Type" )
           print(" 1 - Hourly \n 2 - Day \n 3 - Week")
           rental_type = int(input())
           if rental_type == 1:
               Time = super().hourly()
           elif rental_type == 2:
                Time = super().daily()
           elif rental_type == 3:
                Time = super().weekly()
           else :
                print("Enter Valid number") 
       
                
        
    def return_Cars(self):
        print("Select Your Rental Type" )
        print(" 1 - Hourly \n 2 - Day \n 3 - Week")
        rental_type = int(input())
        if rental_type <= 3 & rental_type > 0 :
            Rental_Amount = super().car_Return(rental_type)
        else :
            print("Enter a Valid Number : ")
        return Rental_Amount;
        
```

```python
def main():
    obj1 = customer(30)
    print("WELCOME TO CARS RENTAL")
    choice=int(input("Kindly enter your choice \n 1 - To rent car \n 2 - To return car \n"))
    if choice == 1:
        car_need = int(input("Enter No of Cars needed :"))
        obj1.request_Cars(car_need)
    elif choice == 2:
        obj1.return_Cars()
    else :
        print("Enter Valid Number")
main()
```
